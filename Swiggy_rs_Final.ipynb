```python
import pandas as pd
data=pd.read_csv(r"swiggy.csv")
display(data.head())
```

```python
# Check the size of the data
display(data.shape)
```

```python
# Remove duplicate rows
data = data.drop_duplicates()

# Re-check after dropping duplicate rows
display(data.shape)
```

```python
# Check null/missing values
data.isnull().sum()
```

```python
# Drop the rows that has null values to improve the analysis
data = data.dropna(subset=['name', 'rating', 'rating_count', 'address','cost','cuisine','lic_no'])

# Check the shape for understanding
display(data.shape)

# Re-check the null values
display(data.isnull().sum())
```

```python
data.info()
```

```python
# Remove unwanted characters and convert to numeric
import pandas as pd

# Clean and convert 'rating'
data['rating'] = pd.to_numeric(data['rating'], errors='coerce').fillna(0)

# Clean and convert 'rating_count'
data['rating_count'] = data['rating_count'].astype(str)

# Remove non-numeric characters (+ and string)

data['rating_count'] = data['rating_count'].str.replace(r'[^0-9]', '', regex=True)

# Convert to numeric
data['rating_count'] = pd.to_numeric(data['rating_count'], errors='coerce').fillna(0).astype(int)

# Convert to string first to avoid errors
data['cost'] = data['cost'].astype(str)

# Remove non-numeric characters (₹ and spaces)
data['cost'] = data['cost'].str.replace(r'[^0-9]', '', regex=True)

# Convert to numeric
data['cost'] = pd.to_numeric(data['cost'], errors='coerce')

# Display data
display(data.head())
```

```python
# Datatype check
data.info()

# Save the cleaned data to a new CSV file
data.to_csv("cleaned_data.csv", index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import pickle

# Load the dataset
df = pd.read_csv('cleaned_data.csv', index_col=0)

# Limit high-cardinality features
def limit_categories(series, top_n):
    top_categories = series.value_counts().nlargest(top_n).index
    return series.where(series.isin(top_categories), other='Other')

df['name'] = limit_categories(df['name'], top_n=500)
df['city'] = limit_categories(df['city'], top_n=100)
df['cuisine'] = limit_categories(df['cuisine'], top_n=200)

# Select columns to encode
categorical_cols = ['name', 'city', 'cuisine']

# Apply One-Hot Encoding
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded = encoder.fit_transform(df[categorical_cols])

# Build DataFrame with encoded features
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(categorical_cols), index=df.index)

# Merge with remaining columns
df_numeric = df.drop(columns=categorical_cols)
final_df = pd.concat([df_numeric, encoded_df], axis=1)

# Save encoder
with open('encoder.pkl', 'wb') as f:
    pickle.dump(encoder, f)

# Save encoded dataset
final_df.to_csv('encoded_data.csv')
```

```python
# Index matching check
import pandas as pd

# Load both the data files
cleaned_df = pd.read_csv("cleaned_data.csv")
encoded_df = pd.read_csv("encoded_data.csv")

# Check row counts
if cleaned_df.shape[0] == encoded_df.shape[0]:
    print("Number of row counts are matching.")
else:
    print("The row counts are not maching")

# Check index match
if (cleaned_df.index == encoded_df.index).all():
    print("Indices are correctly matching.")
else:
    print("Indices are not matching. Re-set again")
    cleaned_df.reset_index(drop=True, inplace=True)
    encoded_df.reset_index(drop=True, inplace=True)
    print("Indices are resetted and matched correctly.")
```

```python
import pandas as pd
import numpy as np
import pickle
from sklearn.cluster import KMeans

# Load encoded data
encoded_data = pd.read_csv("encoded_data.csv")

# Ensure all values are numeric
encoded_data = encoded_data.select_dtypes(include=[np.number])

# Define number of clusters-this cluster value may vary depend on requirement
num_clusters = 5

# Initialize and fit K-Means model
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
encoded_data['Cluster'] = kmeans.fit_predict(encoded_data)

# Save cluster labels
encoded_data.to_csv("clustered_data.csv", index=False)

# Save K-Means model
with open("kmeans_model.pkl", "wb") as f:
    pickle.dump(kmeans, f)
```
